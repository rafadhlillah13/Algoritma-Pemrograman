## Sorting Menggunakan Bubble Sort

```python
def bubbleSort( urt ):
    n = len( urt )
    for i in range( n - 1 ) :
        start = 0
        for j in range(n - 1) :
            if urt[j] > urt[j + 1] :
                tmp = urt[j]
                urt[j] = urt[j + 1]
                urt[j + 1] = tmp
                start = 1
        if start == 0:
            break
    return urt

el = glucose

result = bubbleSort(el)
print("Bubble Sort Of Glucose : \n")
print(result)
```

## Searching DataFrame

```python
# Class searching data untuk dataframe
class Search:
    def __init__(self, data):
        self.data = data
        self.col = []
        for i in self.data:
            self.col.append(i)
    def searching(self, column, item):
        if column in self.col:
            if item not in list(self.data[column]):
                return f"No {item} in Columns {column}"
            else:
                return self.data[self.data[column] == item]
        else:
            return f"No Columns {column} in Data Frame"
```

```python
glucose_s = Search(df)
glucose_s.searching("Glucose", 116)
```
